# Experimento de la doble rendija
Para la elaboración del experimento que demuestra el fenómeno de la doble rendija, el cual ilustra principios básicos de la mecánica cuántica y la naturaleza ondulatoria de la luz, se usaron los siguientes materiales:

- Un puntero láser.
- Varias hojas de papel aluminio.
- Cinta adhesiva.
- Una pantalla o pared blanca.
- Una Caja
- Soporte para el láser.

## Descripción del experimento:

### 1. Preparación de la doble rendija:
   Se toma una hoja de papel aluminio y se dobla por la mitad para darle más rigidez. Con un bisturí o una aguja, se perfora cuidadosamente dos rendijas muy finas y paralelas, con una separación de aproximadamente medio a un milímetro entre ellas. Se debe tener cuidado de que las rendijas sean lo más rectas y similares posible.

### 2. Montaje del experimento:
   Se pega con cinta adhesiva el papel aluminio con las rendijas en un marco o soporte para que quede bien estirado y sin deformaciones. El marco puede ser simplemente el borde de una abertura en una caja de cartón.

### 3. Colocación del láser y pantalla:
   Se usa un soporte para el puntero láser, de manera que quede fijo y apunte directamente a las rendijas en el papel aluminio. La distancia entre el láser y las rendijas puede ser de un par de metros, asegurándote de que la luz incida perpendicularmente sobre las rendijas. Se alinea con una pared o una pantalla blanca detrás de las rendijas para observar el patrón de interferencia.

### 4. Realización del experimento:
   Se enciende el puntero láser y se observa lo que sucede en la pantalla. Se debería ver una serie de franjas claras y oscuras. Estas son el resultado de la interferencia constructiva y destructiva de las ondas de luz que pasan a través de las rendijas, respectivamente.

### 5. Observación y análisis:
   Las franjas iluminadas se deben a que las ondas de luz de cada rendija se suman (interferencia constructiva), mientras que las franjas oscuras ocurren donde las ondas se restan (interferencia destructiva). Si se tapa una de las rendijas, verá cómo cambia el patrón, lo que demuestra que el patrón de interferencia es un fenómeno de ondas superpuestas.

Este experimento casero es una versión simplificada del que realizó Thomas Young en el siglo XIX, y es fundamental para comprender la naturaleza de la luz, actuando a veces como partícula y otras como onda. Este fenómeno también es un punto de partida para discutir cómo funciona la naturaleza en la mecánica cuántica.

## Simulación
A continuación presentamos una simulación del experimento, mostrando en un diagrama la interferencia de onda usando las librerias Numpy y Matplotlb de Python. Los datos usados para ésta simulación son aproximaciones de los datos reales.

```python
import numpy as np
import matplotlib.pyplot as plt

distance_between_slits = 0.1  
wavelength = 0.0006  
distance_to_screen = 2.0  
num_points = 1000  


k = 2 * np.pi / wavelength  

screen_width = 0.1  
screen_positions = np.linspace(-screen_width / 2, screen_width / 2, num_points)

phase_diff = (k * screen_positions * distance_between_slits) / distance_to_screen

intensity_1 = np.cos(phase_diff / 2) ** 2  
intensity_2 = np.cos(phase_diff / 2) ** 2  

total_intensity = intensity_1 + intensity_2

total_intensity /= np.max(total_intensity)

plt.plot(screen_positions, total_intensity, color='blue')
plt.title('Patrón de interferencia de la doble rendija')
plt.xlabel('Posición en la pantalla(metros)')
plt.ylabel(' Intensidad')
plt.ylim(0, 1.1)
plt.show()
```
